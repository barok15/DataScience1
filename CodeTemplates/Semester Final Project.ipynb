```python
#COMBINE PLAYERS + SALARY 
#Remove every column from players except ID and name 

columns_keep = ['_id', 'position', 'name']
for column in players.columns: 
    if column not in columns_keep: 
        players.drop(column, axis = 1, inplace=True)

Salary = pd.concat([salary, players], axis = 1)
```

```python
Salary.head(11)
```

```python
# NEW SALARY DATASET 
#Iterate through season_start column 
#If not >= 2017 remove it 
#Delete the "league" column
#print new dataset 

column_name = 'season_start'
for value in Salary[column_name]:
    index = Salary[Salary[column_name] == value].index[0]
    if value < 2017: 
        Salary = Salary.drop(index)

Salary = Salary.reset_index(drop=True)

del Salary['league'] # Delete league column 
```

```python
nba_stats = nba_stats[nba_stats['Season'] == '2017-18']
nba_stats = nba_stats.reset_index(drop=True)
```

```python
nba_stats.head()
```

```python
total = 0
scorers = []

for index, row in nba_stats.iterrows():
    if row['PTS'] >= 25:
        total += 1
        scorers.append((row['Player'], row['PTS'], row['Pos']))

print("Total high-scorers:", total)

# Print the names, points, and positions of high-scorers
for player, points, position in scorers:
    print(f"Name: {player}, Points: {points}, Position: {position}")
```
